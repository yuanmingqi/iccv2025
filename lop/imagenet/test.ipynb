```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
from lop.utils.miscellaneous import *
```

```python
all_test_accuracies = []
for seed in range(2):
    data = pickle.load(open(f'data/ent_cbp/0/{seed}', 'rb'))
    print(seed, data['test_accuracies'].shape)
    acc = np.array(bin_m_errs(errs=data['test_accuracies'][:1000, -1].flatten()*100, m=50)) 
    all_test_accuracies.append(acc)

all_test_accuracies = np.array(all_test_accuracies)
mean_test_accuracies = np.mean(all_test_accuracies, axis=0)
std_test_accuracies = np.std(all_test_accuracies, axis=0)

plt.plot(mean_test_accuracies, label='Test accuracy')
plt.fill_between(range(len(mean_test_accuracies)), mean_test_accuracies-std_test_accuracies, mean_test_accuracies+std_test_accuracies, alpha=0.3)
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.legend()
```
